```python
import pandas as pd
```

```python
df=pd.read_csv("/content/HR Data.csv")

df.head(5)
```

```python
df.describe()
```

```python
# List of columns to remove
columns_to_remove = ['EmployeeCount', 'EmployeeNumber', 'StandardHours']

# Drop the unnecessary columns
df = df.drop(columns=columns_to_remove)
```

```python
df.head(5)
```

```python
# Dictionary mapping old column names to new column names
new_column_names = {
    'Age': 'Employee_Age',
    'Attrition': 'Attrition_Status',
    'BusinessTravel': 'Travel_Frequency',
    'DailyRate': 'Daily_Rate',
    'Department': 'Department_Type',
    'DistanceFromHome': 'Distance_From_Home',
    'Education': 'Education_Level',
    'EducationField': 'Education_Field',
    'EnvironmentSatisfaction': 'Environment_Satisfaction',
    # Add more column name mappings as needed
}

# Rename the columns
df = df.rename(columns=new_column_names)
```

```python
df.head(5)
```

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

```python
df.head(5)
```

```python
# Sanitize specific columns
# For example, let's remove leading and trailing whitespaces from the 'BusinessTravel' column
df['Travel_Frequency'] = df['Travel_Frequency'].str.strip()

# Convert 'DailyRate' column to integer data type
df['Daily_Rate'] = df['Daily_Rate'].astype(int)

# Replace missing values in 'Education' column with the median
df['Education_Level'].fillna(df['Education_Level'].median(), inplace=True)
```

```python
df.head(5)
```

```python
# Define the columns you want to sanitize (remove NaN values)
columns_to_sanitize = ['Employee_Age', 'Attrition_Status', 'Travel_Frequency', 'Education_Field', 'Department_Type']

# Remove NaN values from the specified columns
df.dropna(subset=columns_to_sanitize, inplace=True)

# Display the modified DataFrame
print(df.head())
```
